### 17. Build a dataset with every possible (UserID, ProductID) pair (cross join) 

```python
# There are no common keys on users and products data frame to merge.
# create a new data frame with column key and userid.
# add value 1 for each row in data frame for column key.
df1 = pd.DataFrame({'key':np.repeat(1, users.shape[0]), 'UserID': users.UserID})

# create new data frame with column key and productid.
# add value 1 for each row in data frame for column key.
df2 = pd.DataFrame({'key':np.repeat(1, products.shape[0]), 'ProductID': products.ProductID})

# merge 2 data frames on column 'key' which are created above to get the result of cross join.
df1.merge(df2, on='key')[['UserID','ProductID']]
```

### 18. Determine how much quantity of each product was purchased by each user 

```python
# There are no common keys on users and products data frame to merge.
# create a new data frame with column key and userid.
# add value 1 for each row in data frame for column key.
df1 = pd.DataFrame({'key':np.repeat(1, users.shape[0]), 'UserID': users.UserID})

# create new data frame with column key and productid.
# add value 1 for each row in data frame for column key.
df2 = pd.DataFrame({'key':np.repeat(1, products.shape[0]), 'ProductID': products.ProductID})

# merge 2 data frames on column 'key' which are created above to get the result of cross join. Select only 2 columns.
s = df1.merge(df2, on='key')[['UserID','ProductID']]

# merge the new data frame with transactions using left join on UserId and ProductID columns.
# Group the resulted data frame with 'UserID' and 'ProductID' columns.
# for each group of user id and product id, get the sum of quantity for each group.
# Reset the index of newly created dataframe with few columns.
# fill zero value to all the cell which contains value as NaN/na/null.
pd.merge(s, transactions, how='left', on=['UserID','ProductID']).groupby(['UserID','ProductID']).apply(
    lambda e: pd.Series(dict(Quantity=e.Quantity.sum()))).reset_index().fillna(0)
```

### 25. Get the name of the product which was maximum purchased.

```python
# Group transactions datafrane by ProductID column.
t = transactions.groupby('ProductID')
# Create a series with sum of Quantity of each product. Apply this function to the above groupby object created.
t = t.apply(lambda e: pd.Series(dict(Quantity=e.Quantity.sum())))
# Merge above created dataframe with products table using left join on ProductID column.
t = t.merge(products, how='left', on='ProductID')
# Add filter to select maximum quantity of the product purchased.
t[t.Quantity == max(t.Quantity)]
```

### 27. Get the name of the user who has attended maximum session.

```python
# Group sessions dataframe by UserID and get the session count of each user. 
s = sessions.groupby('UserID').count()
# Merge above data frame with users on UserID column using left merge.
s = s.merge(users, on='UserID', how='left')
# Select the user details who attended maximum number of sessions.
s = s[s.SessionID == max(s.SessionID)]
# print the user name.
print(s.User)
```

### 28. Name the user who spent least amount

```python
# Merge the transaction data frame with products to get the price details.
a = transactions.merge(products, on='ProductID')
# Group above data frame by UserID.
a = a.groupby('UserID')
# Create a series with sum of Price of each user spent. Apply this function to the above groupby object created.
a = a.apply(lambda e: pd.Series(dict(Price = e.Price.sum())))
# Merge the above data frame with users data frame to get the user name.
a = a.merge(users, on='UserID')
# Add condition to select the user who spend less amount.
a = a[a.Price == min(a.Price)]
# print the user name.
print(a.User)
```
